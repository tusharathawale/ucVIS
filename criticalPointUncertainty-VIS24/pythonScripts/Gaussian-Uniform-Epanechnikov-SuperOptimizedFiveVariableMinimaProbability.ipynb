```python
def add_Gaussian_noise(field, noise_level, numMembers):
    h,w = field.shape
    ensemble = np.zeros((h,w,numMembers))
    
    for i in range(numMembers):
        epsilon = np.random.normal(0,noise_level,[h,w])
        ensemble[:,:,i] = field + epsilon
        
    return ensemble
```
